```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('/content/Life Expectancy at Birth.csv')  # Replace with your real file path if different
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(
    data=df,
    y='Life Expectancy at Birth (2021)',
    bins=20,
    color='skyblue'
)
plt.title('Life Expectancy at Birth (2021)')
plt.xlabel('Frequency')
plt.ylabel('Life Expectancy')
plt.tight_layout()
plt.show()
```

Horizontal Histogram

```python
sns.displot(
    df,
    x='Life Expectancy at Birth (2021)',
    bins=15,
    hue='Continent',
    multiple='stack'
).set(
    title='Life Expectancy by Continent',
    xlabel='Life Expectancy at Birth (2021)',
    ylabel='Frequency'
)
plt.show()
```

Distribution Plots With hue by Continent

```python
sns.displot(
    df,
    x='Life Expectancy at Birth (2021)',
    bins=15,
    col='Hemisphere'
).set(
    title='Life Expectancy by Hemisphere',
    xlabel='Life Expectancy at Birth (2021)',
    ylabel='Frequency'
)
plt.show()
```

 Distribution Split by Hemisphere

**2D Histograms and KDE**

```python
sns.displot(
    df,
    x='Life Expectancy at Birth (2021)',
    y='Life Expectancy at Birth (1990)',
    cbar=True
).set(
    title='2021 vs 1990 Life Expectancy',
    xlabel='2021',
    ylabel='1990'
)
plt.show()
```

Histogram + Color Bar

```python
sns.displot(
    df,
    x='Life Expectancy at Birth (2021)',
    y='Life Expectancy at Birth (1990)',
    kind='kde',
    rug=True
).set(
    title='KDE: 2021 vs 1990 Life Expectancy',
    xlabel='2021',
    ylabel='1990'
)
plt.show()
```

 KDE + Rug


```python
g = sns.JointGrid(
    data=df,
    x='Life Expectancy at Birth (2021)',
    y='Life Expectancy at Birth (1990)'
)
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
plt.show()
```

JointGrid Visualization

**Categorical Plots**

 Boxen Plot

```python
sns.catplot(
    data=df,
    x='Continent',
    y='Life Expectancy at Birth (2021)',
    hue='Hemisphere',
    kind='boxen'
).set(
    title='Life Expectancy by Continent and Hemisphere',
    xlabel='Continent',
    ylabel='Life Expectancy (2021)'
)
plt.show()
```

```python
sns.catplot(
    data=df,
    x='Continent',
    y='Life Expectancy at Birth (2021)',
    hue='Hemisphere',
    kind='box'
).set(
    title='Box Plot: Life Expectancy (2021)',
    xlabel='Continent',
    ylabel='Life Expectancy'
)
plt.show()
```

 Box Plot

```python
sns.catplot(
    data=df,
    x='Continent',
    y='Life Expectancy at Birth (2021)',
    hue='Hemisphere',
    kind='violin'
).set(
    title='Violin Plot: Life Expectancy (2021)',
    xlabel='Continent',
    ylabel='Life Expectancy'
)
plt.show()
```

Violin Plot

```python
sns.catplot(
    data=df,
    x='Continent',
    y='Life Expectancy at Birth (2021)',
    hue='Hemisphere',
    kind='violin',
    split=True
).set(
    title='Split Violin Plot: Life Expectancy (2021)',
    xlabel='Continent',
    ylabel='Life Expectancy'
)
plt.show()
```

Violin Plot Split

```python
sns.catplot(
    data=df,
    x='Continent',
    y='Life Expectancy at Birth (2021)',
    hue='Hemisphere',
    kind='point'
).set(
    title='Point Plot: Life Expectancy (2021)',
    xlabel='Continent',
    ylabel='Life Expectancy'
)
plt.show()
```

Point Plot

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

sns.histplot(
    df['Life Expectancy at Birth (1990)'],
    bins=15,
    ax=ax1,
    color='coral'
)
ax1.set_title('Life Expectancy in 1990')
ax1.set_xlabel('Life Expectancy')
ax1.set_ylabel('Count')

sns.histplot(
    df['Life Expectancy at Birth (2021)'],
    bins=15,
    ax=ax2,
    color='seagreen'
)
ax2.set_title('Life Expectancy in 2021')
ax2.set_xlabel('Life Expectancy')
ax2.set_ylabel('Count')

plt.tight_layout()
plt.show()
```

Subplots: Two Distribution Plots Side-by-Side

```python
# Selecting only Life Expectancy columns
life_exp_cols = [col for col in df.columns if 'Life Expectancy' in col]
corr = df[life_exp_cols].corr()

plt.figure(figsize=(14, 10))
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title('Correlation Heatmap of Life Expectancy (1990–2021)')
plt.show()
```

Heatmap of Correlations Between Life Expectancy Years

```python
selected_years = [
    'Life Expectancy at Birth (1990)',
    'Life Expectancy at Birth (2000)',
    'Life Expectancy at Birth (2010)',
    'Life Expectancy at Birth (2021)'
]

sns.pairplot(df[selected_years])
plt.suptitle('Pair Plot of Life Expectancy Over Years', y=1.02)
plt.show()
```
