###1. What is the difference between a spreadsheet and a database?

*  Spreadsheets (Microsoft Excel, Google Sheets) are tools designed for organizing and analyzing data in a flat, tabular format using rows and columns. They are suitable for small datasets, allowing users to perform basic calculations, sorting, filtering, and visualization using built-in tools.
*  Databases (Microsoft SQL Server) are designed to handle large and complex datasets efficiently. They store data in multiple structured tables with defined relationships, ensuring data integrity, scalability, and multi-user access. Unlike spreadsheets, databases enforce data normalization to reduce redundancy and maintain consistency.



###2. What are the advantages and disadvantages of using a spreadsheet for health data management?

*   Advantages:
  *   User-friendly interface – Easy to use for individuals with minimal technical knowledge.
  *  Simple data entry and formatting – Allows manual data input and customization with formatting options.
  *   Basic data analysis and visualization – Provides built-in functions for calculations, creating charts, and summarizing data using pivot tables.
  *   Flexibility – Can be used for quick, small-scale data manipulation and reporting.
  *   Compatibility – Can be exported to other formats (e.g., CSV, XLSX) for further analysis in software like R or Python.

*   Disadvantages:
  *   Limited scalability – Not ideal for handling large datasets, as performance may degrade with increased data volume.
  *   Data integrity issues – No built-in mechanisms to prevent duplicate or inconsistent data entry.
  *   Lack of multi-user access – Difficult for multiple users to edit simultaneously without conflicts.
  *   Security limitations – Basic protection options, but lacks robust access controls like user authentication and encryption.
  *   Error-prone – Manual data entry increases the likelihood of errors and inconsistencies.


###3. What are the advantages and disadvantages of using a database for health data management?

*   Advantages:
    * Handles large and complex datasets efficiently.
    * Ensures data integrity and supports relationships between data elements.
    * Allows multiple users to access and manipulate data simultaneously.
    * Provides advanced security and access controls.
*   Disadvantages:
    * Requires knowledge of database design and SQL.
    * More complex to set up and maintain compared to spreadsheets.

###8.  What are some examples of tasks that can be performed using a spreadsheet but not a database?
*  Quick manual calculations – Simple arithmetic functions like SUM, AVERAGE, and COUNT.
* Ad-hoc data entry – No predefined structure required, allowing for flexible data input.
* Graphical customization – Formatting cells with colors, bold text, and other stylistic adjustments.
* On-the-fly data visualization – Creating charts without needing external software.
* Quick one-off data manipulations – Temporary sorting, filtering, or reorganizing without affecting stored records.

###9.  What are some examples of tasks that can be performed using a database but not a spreadsheet?
* Managing relational data – Storing patient-doctor relationships across multiple tables.
* Enforcing complex data constraints – Ensuring unique patient IDs, preventing duplicate records.
* Handling large-scale healthcare systems – Supporting data access for hospitals and clinics.
* Processing concurrent transactions – Allowing multiple users to update patient records simultaneously.
* Executing complex SQL queries – Retrieving insights from thousands of records efficiently.
